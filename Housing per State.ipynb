```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("../City_Formatted1.csv")
city_df = pd.DataFrame(data)
city_df
```

```python
top = 360000
bot = 150000
rtop = 1900
rbot = 1000
xtic = ["9/30/2014","12/31/2014","3/31/2015","6/30/2015","9/30/2015","12/31/2015","3/31/2016","6/30/2016",
        "9/30/2016","12/31/2016","3/31/2017","6/30/2017","9/30/2017","12/31/2017","3/31/2018","6/30/2018",
        "9/30/2018","12/31/2018","3/31/2019","6/30/2019"]
```

```python
nash = city_df.loc[city_df["City"] == "Nashville", :]

nash_plot_all, = plt.plot(nash["Date"], nash["Median Listing Price: All Homes"], c="black", label ="All Homes")
nash_plot_single, = plt.plot(nash["Date"], nash["Median Listing Price: Single Family"], c="blue", label ="Single Family")
nash_plot_condo, = plt.plot(nash["Date"], nash["Median Listing Price: Condos"], c="red", label ="Condos")

nash_x = np.arange(0, 58)
nash_y = nash['Median Listing Price: All Homes']

nash_z = np.polyfit(nash_x, nash_y, 1)
nash_p = np.poly1d(nash_z)
plt.plot(nash_x,nash_p(nash_x),"g--")

plt.legend(handles=[nash_plot_all,nash_plot_single,nash_plot_condo],loc="lower right")
plt.grid(axis = "y")
plt.title('Housing Cost: Nashville')
plt.xlabel('Date')
plt.ylabel('Housing Price')
plt.ylim(bot,top)
plt.xticks(xtic,rotation=55,ha="right")
plt.tight_layout()

plt.savefig("housing_nashville.png")
```
